# Examination 2024-2025 

**Goal**: Identify AGNs in Euclid data based on catalog information, visualise the candidate in a Euclid image, and evaluate how many sources happen to be in projection within 2 arcsec of each AGN candidate. 

**Data** 
You are provided with one Euclid `fits` image, three catalogs, and a `png` file. 

Two catalogs with photometric informations (and a few more info., see below) for stars and AGNs: 
- `Catalog_stars.fits`: A fits table catalog of objects identified as stars in Euclid. 
- `Catalog_AGN.fits`: A fits table catalog of objects identified as AGN in Euclid 

One Euclid image (in VIS band) and an associated catalog: 
- `EUC_MER_BGSUB-MOSAIC-VIS_TILEXXXXXXXXXX_sub.fits`: a 5000x5000 pixels cutout of an image observed by Euclid. 
- `MER_XXXXXXXXXXX.fits`: A fits table containing photometric data for the image (called TILE) XXXXXXXXXX. 
This catalog covers a field larger than your image. 
- `Image.png`: A figure showing the image (this is just to get an idea of the image, you do not have to use it!)

The tables contain several information. Some are relevant, others are not directly useful for the exercise.  The most important information are the id of the object, their coordinate, and their photometry. There are also some flags and and some morphological information available. 

The photometry is provided for each source in multiple filters: 
- vis: Euclid VIS filter
- y, j, h: Euclid y, j, h filters (near infra-red) 
- nir_stack: Euclid "combined" near infra-red photometry
- u_ext_megacam, r_ext_megacam: Ground based u and r- band photometry from MEGACAM camera
- i_ext_panstarrs: Ground based i-band photometry from Panstarr survey
- g_ext_hsc, z_ext_hsc: Ground based  g and z photometry from HSC (Subaru telescope)

Brief description of important columns of the table: 
- 'object_id': ID of the object as detected in Euclid images.
- 'gaia_id': Id of object in Gaia catalog (when the object has been detected by Gaia)
- "flux_filter_4fwhm_aper": Total flux of the object in a given filter (filter list above)
- "fluxerr_filter_4fwhm_aper": Uncertainty on the total flux. 
- "flux_detection_total": Total flux in Euclid VIS and NIR
- 'vis_det': 0 if detected in a Euclid VIS image 
- 'point_like_prob': Probability for the source to be a point-source
- 'det_quality_flag': Flag indicative of the quality of the entry. DET_QUALITY_FLAG < 4 are best. 

**Tasks** 

- Task 1: Use the catalogs `Catalog_stars.fits` and `Catalog_AGN.fits` to train (one or several) 
classifier(s) that identify AGNs among point-like sources based on their photometry. 

- Task 2: Apply your trained classifier on the Euclid catalog MER_XXXXXXXXXXX.fits and make a list of AGN candidates.  
Because the catalog also contains galaxies and spurious object, 
you should first filter your catalog based on the following criteria: 
    * `det_quality_flag` < 4 # less problematic sources   
    * `flag_vis` = 0 
    * `flux_detection_total` / `fluxerr_detection_total` > 12  (<-> SNR > 12)
    * `point_like_prob` > 0.8
    * `flux_detection_total` > 3.13   (This is the same flux limit as the star catalog)
- Task 3: Identify which of the AGNs found at step 2 are visible in your image. 
Make a figure that displays your 10 best candidates (consider image stamps of 100 x 100 pixels). If you do not have any candidate (or less than 3 candidates falling in the footprint of your image), you can display 10 stars. 
- Task 4: Detect and count all the sources falling within a radius of 2 arcsec of each of your AGN candidate (present in the footprint of your image !). If you have less than 3 candidates, do the same for 10 stars. 

**TIP** 
The tables contain fluxes in micro-Jy. Magnitudes may be a wiser choice of feature. 
One can convert fluxes to magnitudes using the formula given below. 
**To convert fluxes to magnitudes**:  
Use the following formula:  $ -2.5*log10(flux) + 23.9$
where flux is the flux in one of the photometric column.
